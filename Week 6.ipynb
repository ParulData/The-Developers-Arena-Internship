```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulating Sales Data
np.random.seed(0)
n_customers = 30
data = {
    'CustomerID': [f'CUST_{i+1}' for i in range(n_customers)],
    'Region': np.random.choice(['North', 'South', 'East', 'West'], n_customers),
    'Segment': np.random.choice(['Consumer', 'Corporate', 'Home Office'], n_customers),
    'Total_Purchases': np.random.randint(5, 50, n_customers),
    'Average_Purchase_Value': np.random.uniform(20, 200, n_customers),
}
df = pd.DataFrame(data)
df
```

```python
df['Total_Spent'] = df['Total_Purchases'] * df['Average_Purchase_Value']

# Pivot Table: Summarize Total Spent by Region and Segment
pivot = df.pivot_table(index='Region', columns='Segment', values='Total_Spent', aggfunc='sum')
print("\nPivot Table (Total Spent by Region & Segment):")
print(pivot)

# Seaborn Chart 1: Boxplot for Total_Spent per Region
plt.figure(figsize=(8,5))
sns.boxplot(x='Region', y='Total_Spent', data=df)
plt.title('Total Spent Distribution by Region')
plt.show()

# Seaborn Chart 2: Heatmap for Pivot Table summary
plt.figure(figsize=(8,5))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title('Total Sales by Region and Segment (Heatmap)')
plt.show()
```
