```python
def grade_calculator(percentage):
    if percentage<= 100 and percentage >= 90:
        print("Final Grade : A")
        print("Excellent work! You’ve really mastered the concepts.")
    elif percentage<= 90 and percentage >= 80:
        print("Final Grade : B")
        print("Great job! You’re doing very well—keep pushing for that A!")
    elif percentage<= 80 and percentage >= 70:
        print("Final Grade : C")
        print("Good effort! A bit more consistency and you’ll reach higher grades.")
    elif percentage<= 70 and percentage >= 60:
        print("Final Grade : D")
        print("You passed! Keep working steadily—you’re on the right path.")
    elif percentage<= 60 and percentage >= 50:
        print("Final Grade : E")
        print("You’re almost there! Review your weak areas and you’ll improve fast.")
    else:
        print("Final Grade : F")
        print("Don’t be discouraged! Every failure is a step toward success. Learn and try again!")


Science = float(input("Enter Science marks : "))
Chemistry = float(input("Enter Chemistry marks : "))
Maths = float(input("Enter Maths marks : "))

percentage = ((Science + Chemistry + Maths) / 300) * 100

print("Total Percentage : ", percentage)
grade_calculator(percentage)
```
