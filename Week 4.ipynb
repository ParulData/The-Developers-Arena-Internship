```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Simulate data
np.random.seed(10)
n_students = 50
student_ids = [f'Student_{i+1}' for i in range(n_students)]
subjects = ['Math', 'Science', 'English', 'History']

data = {
    'StudentID': student_ids,
    'Math': np.random.randint(50, 100, n_students),
    'Science': np.random.randint(55, 98, n_students),
    'English': np.random.randint(45, 95, n_students),
    'History': np.random.randint(50, 100, n_students),
    'Attendance (%)': np.round(np.random.uniform(65, 100, n_students), 1)
}
df = pd.DataFrame(data)
df['Average'] = df[subjects].mean(axis=1)

# Chart 1: Average subject scores (Bar Plot)
subject_avg = df[subjects].mean()
plt.figure(figsize=(8,5))
plt.bar(subjects, subject_avg)
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.title('Average Scores by Subject')
plt.show()

# Chart 2: Distribution of students average scores (Histogram)
plt.figure(figsize=(8,5))
plt.hist(df['Average'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Average Score')
plt.ylabel('Number of Students')
plt.title("Distribution of Students' Average Scores")
plt.show()

# Written insights
print("Insights:")
print(f"Subject Averages: {subject_avg.to_dict()}")
print(f"Average Attendance: {df['Attendance (%)'].mean():.2f}%")
print("Top 5 Students by Average Score:")
print(df.sort_values('Average', ascending=False)[['StudentID', 'Average']].head(5))
```
