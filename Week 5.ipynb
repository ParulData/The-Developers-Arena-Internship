```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate Sales Data
np.random.seed(0)
n_customers = 30
data = {
    'CustomerID': [f'CUST_{i+1}' for i in range(n_customers)],
    'Region': np.random.choice(['North', 'South', 'East', 'West'], n_customers),
    'Total_Purchases': np.random.randint(5, 50, n_customers),
    'Average_Purchase_Value': np.random.uniform(20, 200, n_customers),
}
df = pd.DataFrame(data)
df['Total_Spent'] = df['Total_Purchases'] * df['Average_Purchase_Value']

# Top 5 Customers by Total Spent
top_customers = df.sort_values('Total_Spent', ascending=False).head(5)
print("Top 5 Customers by Spend:")
print(top_customers[['CustomerID', 'Region', 'Total_Spent']])

# Pivot Table: Summarize Total Spent by Region
pivot = df.pivot_table(index='Region', values='Total_Spent', aggfunc=['sum', 'mean', 'max'])
print("\nPivot Table (Total Spent summary by Region):")
print(pivot)

# Chart 1: Bar plot of Top Customers
plt.figure(figsize=(8,5))
plt.bar(top_customers['CustomerID'], top_customers['Total_Spent'], color='purple')
plt.xlabel('Customer')
plt.ylabel('Total Spent')
plt.title('Top 5 Customers by Spend')
plt.show()

# Chart 2: Histogram of Purchase Value
plt.figure(figsize=(8,5))
plt.hist(df['Average_Purchase_Value'], bins=10, color='orange', edgecolor='black')
plt.xlabel('Average Purchase Value')
plt.ylabel('Number of Customers')
plt.title('Distribution of Average Purchase Value')
plt.show()
```
